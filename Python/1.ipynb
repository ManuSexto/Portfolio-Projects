```python
# Construimos la 2ª alternativa de algoritmos para resolver la identificación de la letra de un DNI y
# Calculamos la letra que corresponde con el Numero del DNI, aplicando la Función:
def calcular_letra(dni):
    letras = "TRWAGMYFPDXBNJZSQVHLCKE"
    return letras[int(dni) % 23]

# Convertimos la columna 'Numero' a int (por si no lo es)
df_dni8digitos['Numero'] = df_dni8digitos['Numero'].astype(int)

# Creamos una columna vacía para las letras calculadas
df_dni8digitos['letra_calculada'] = ''

# Calcular la letra para cada fila manualmente
for index, row in df_dni8digitos.iterrows():
    numero_dni = row['Numero']
    letra_calculada = calcular_letra(numero_dni)
    df_dni8digitos.at[index, 'letra_calculada'] = letra_calculada

# Comparamos con la letra original
df_dni8digitos['Comparación'] = df_dni8digitos['Letra'] == df_dni8digitos['letra_calculada']

print(df_dni8digitos)
```

```python
# Contamos el número total de DNIs en el dataframe original: df_original
total_dnis = len(df_original)
print("El nº total de DNIs del archivo: reto_dni.csv = ", total_dnis)
```

También el número de intentos máximo será una constante en la cabecera del Script.


```python
# Simulación para el caso de introducir por patalla el nº DNI, con un límite de 3 intentos:
def calcular_letra_dni(numero):
    letras_dni = "TRWAGMYFPDXBNJZSQVHLCKE"
    return letras_dni[numero % 23]

def validar_dni(dni):
    if len(dni) != 9:
        return False
    numero, letra = dni[:-1], dni[-1]
    if not numero.isdigit() or letra.upper() != calcular_letra_dni(int(numero)):
        return False
    return True

intentos_maximos = 3
for intento in range(1, intentos_maximos + 1):
    dni = input("Introduce tu nº DNI (formato 8 dígitos y 1 letra): ")
    if validar_dni(dni):
        print("El nº DNI introducido es correcto.")
        break
    else:
        print(f"El nº DNI introducido no es correcto, prueba otra vez, te quedan {intentos_maximos - intento} intentos.")
        if intento == intentos_maximos:
            print("Has alcanzado el número máximo de intentos.")

```
